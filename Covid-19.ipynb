```python
b=a.drop(columns=['New deaths', 'New recovered', 'Deaths / 100 Cases', 'Recovered / 100 Cases', 'Deaths / 100 Recovered', 'Confirmed last week', '1 week change', '1 week change', '1 week % increase','WHO Region'], axis=0)
b
```

```python
b.style.background_gradient(cmap="cubehelix")
```

```python
c=b.groupby(by='Country/Region').max()[['Confirmed']].sort_values(by=['Confirmed'], ascending=False).reset_index()
```

```python
c.head(10)
```

```python
fig=plt.figure(figsize=(16,9))
plt.title("Top 10")
d=sns.barplot(data=c.iloc[:10], y="Confirmed", x="Country/Region")
plt.xlabel("Country")
plt.ylabel("cases")
plt.show()
```

```python
e=b.groupby(by='Country/Region').max()[['Deaths']].sort_values(by=['Deaths'], ascending=False).reset_index()
```

```python
fig=plt.figure(figsize=(16,9))
plt.title("Top 10")
d=sns.barplot(data=e.iloc[:10], y="Deaths", x="Country/Region")
plt.xlabel("Country")
plt.ylabel("cases")
plt.show()
```

```python
fig=plt.figure(figsize=(16,9))
plt.title("Top 10")
f=sns.lineplot(data=e.iloc[:10], y="Deaths", x="Country/Region")
plt.xlabel("Country")
plt.ylabel("cases")
plt.show()
```
